# AI Resume Screening Demo

This notebook demonstrates the modular LangChain pipeline for resume screening, including skill extraction, matching, scoring, and explanation. It runs three candidate examples: strong, average, and weak.

```python
import os
from dotenv import load_dotenv
load_dotenv()
os.environ.setdefault('LANGCHAIN_TRACING_V2', 'true')

from main import ResumeScreeningPipeline, create_llm, build_tracer, load_examples
```

```python
tracer = build_tracer()
llm = create_llm()
pipeline = ResumeScreeningPipeline(llm=llm, tracer=tracer)
resumes, job_description = load_examples()

results = []
for candidate in resumes:
    result = pipeline.run_candidate(candidate, job_description)
    results.append(result)

results
```
